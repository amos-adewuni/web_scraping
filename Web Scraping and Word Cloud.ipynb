###  Web Scraping and WordCloud of President Buhari 2022 Budget Presentation Speech

President Muhammadu Buhari speech during the presentation of the Federal Budget Proposals to the Joint Session of the National Assembly.

Import requests and BeautifulSoup libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
html_text = requests.get("https://statehouse.gov.ng/news/2022-budget-speech-budget-of-economic-growth-and-sustainability-president-buhari/").text
```

```python
soup = BeautifulSoup(html_text, "lxml")
speech = soup.find('div', class_ = 'article__body post-content').text.replace('\n','')
speech
```

```python
#Calculating the number of words and characters in the speech

word_list = speech.split()
number_of_words = len(word_list)
print(f"There are {number_of_words} number of words in the President's Speech")
```

```python
characters = len(speech)
print(f"There are {characters} in the President's Speech")
```

```python
# import WordCloud and StOPWORD modules

from wordcloud import WordCloud, STOPWORDS
```

```python
# Generate the wordcloud object

stopwords = set(STOPWORDS)
speech_wc = WordCloud(background_color = 'white').generate(speech)
```

```python
# import matplotlib for visualising the wordcloud object

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (16,12))
plt.imshow(speech_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show
```
